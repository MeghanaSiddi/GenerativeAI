## Prompt Templating

```python
from utils import get_model_from_gcp
llm=get_model_from_gcp()
```

```python
# I want to prompt to llm to find capital city of a dynamic coubtry

# python based approach

country=input("Enter the country to know capital city:")
prompt=f"What is the capital of {country}"
response=llm.invoke([prompt])
response.pretty_print()
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate.from_template(
    template="what is the capital of {country}? and also show their achievements in the following sport {sport}"
)

chain = prompt_template|llm
response = chain.invoke({
    "country":"India",
    "sport":"Hockey"
})
print(response)
response.pretty_print()

response=chain.invoke(
    {
        "country":"Portugal",
         "sport":"Football"
    }
)
print("---------")

response.pretty_print()
```

```python
from langchain_core.prompts import ChatPromptTemplate

# prompt with roles

prompt= ChatPromptTemplate.from_messages([
    ("system","You are an expert in {topic}."),
    ("user","what is the significance of {topic} for {audience}")
])

chain=prompt|llm
response=chain.invoke({
    "topic":"sports",
    "audience":"students"
})

response.pretty_print()
```

```python
from langchain_core.output_parsers import StrOutputParser

output = StrOutputParser()
chain=prompt|llm|output

response=chain.invoke({
    "topic":"Sports",
    "audience":"students"
})

print(response)
# you can't use response.content here as output is plain text
```

```python
from pydantic import BaseModel,Field

class SupportTicket(BaseModel):
    issue_category:str=Field(description="One of: billing,technical,shipping,others")
    resolved:bool=Field(description="Whether the issue appears resolved based on the conversation")
    next_action:str=Field(description="What should happen next, one short sentence")


prompt=ChatPromptTemplate.from_messages([
    ("system","you summarize customer support conversations into structured tickets."),
    ("user","My last order never arrived and it's been 2 weeks"),
    ("assistant", "I'm sorry to hear that, I have issued a reshipment, tracking number sent to your registred email and phone number"),
    ("user","Got it! Thank you!!")
])

chain=prompt|llm.with_structured_output(SupportTicket,method="json_schema")
response=chain.invoke({})
print(response)
response
```
